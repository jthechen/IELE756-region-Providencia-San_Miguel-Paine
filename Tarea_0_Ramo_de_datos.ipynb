```python
import pandas as pd
import zipfile
import matplotlib.pyplot as plt

# 1. Arreglamos el warning creando una oficial de tu tabla filtrada (crea un Dataframe independiente)
grd_region = grd_region.copy()

# Usamos la ruta de tu Drive (asegúrate de que el nombre de la carpeta sea exacto)
ruta_zip = "/content/drive/MyDrive/Pre. y Análisis de datos/EGRESOS_2020.zip"

with zipfile.ZipFile(ruta_zip) as z:
    with z.open(z.namelist()[0]) as f:
        grd = pd.read_csv(f, sep=";", encoding="latin-1", low_memory=False)

    with z.open(nombre_diccionario) as f:
        # Cargamos el archivo completo para ver sus pestañas
        xls = pd.ExcelFile(f)
        print(f"Pestañas encontradas en el Excel: {xls.sheet_names}")

        # Buscamos la pestaña que tenga "CIE" en su nombre, si no, tomamos la segunda (índice 1)
        hojas_cie = [h for h in xls.sheet_names if "CIE" in h.upper()]
        hoja_correcta = hojas_cie[0] if hojas_cie else xls.sheet_names[1]

        print(f" Leyendo automáticamente la pestaña: '{hoja_correcta}'")

        # Leemos la pestaña correcta saltando los logos (header=8)
        cie10 = pd.read_excel(xls, sheet_name=hoja_correcta, header=8)

        # Limpiamos los nombres de las columnas
        cie10.columns = cie10.columns.astype(str).str.strip().str.upper()

# 2. Verificamos que la columna exista y cruzamos
if "CODIGO SUBCATEGORIA" in cie10.columns:
    print(" ¡Columna encontrada! Haciendo el cruce...")

    # Limpiamos los datos
    grd_region["DIAG1"] = grd_region["DIAG1"].astype(str).str.strip().str.upper()
    cie10["CODIGO SUBCATEGORIA"] = cie10["CODIGO SUBCATEGORIA"].astype(str).str.strip().str.upper()

    # Hacemos el Merge (combinación de data frames)
    grd_region = grd_region.merge(
        cie10[["CODIGO SUBCATEGORIA", "GLOSA SUBCATEGORIA"]],
        left_on="DIAG1",
        right_on="CODIGO SUBCATEGORIA",
        how="left"
    )

    # 3. CREAMOS EL GRÁFICO
    top_5_final = grd_region["GLOSA SUBCATEGORIA"].value_counts().head(5)

    plt.figure(figsize=(12, 6))
    top_5_final.plot(kind="barh", color="mediumseagreen").invert_yaxis()
    plt.title("Top 5 Diagnósticos Hospitalarios - Región de Los Lagos (2020)", fontsize=14)
    plt.xlabel("Cantidad de Egresos", fontsize=12)
    plt.ylabel("Diagnóstico", fontsize=12)
    plt.tight_layout()
    plt.show()

else:
    print(" Aún no encontramos la columna. Las columnas en esta pestaña son:")
    print(cie10.columns.tolist())
```
